```python
import pandas as pd
import numpy as np
import math
from datetime import date

tmrw="May 4 2020"
today="May 3 2020"
today = date.today()
DATE = today.strftime("%Y-%m-%d")

record= pd.read_csv('Output.csv')
yesterday = record.iloc[-1]

ACTUAL_USA = input("USA ACTUAL ")
ACTUAL_USA_DEAD = input("USA Dead ")
USA_DEATH_RATE = int(ACTUAL_USA_DEAD)/int(ACTUAL_USA)
USA_DEATH_RATE = float("{:.5f}".format(USA_DEATH_RATE))
USA_YTD_ACTUAL = int(yesterday['USA ACTUAL'])
USA_RATE_OF_CHANGE1= int(ACTUAL_USA)-int(yesterday['USA ACTUAL'])
USA_NAT_LOG = np.log(int(ACTUAL_USA)/int(USA_YTD_ACTUAL))
USA_NAT_LOG = float("{:.6f}".format(USA_NAT_LOG))


ACTUAL_CANADA = input("Canada Actual ")
ACTUAL_CAN_DEAD = input("Canada Dead ")
CAN_DEATH_RATE = int(ACTUAL_CAN_DEAD)/int(ACTUAL_CANADA)
CAN_DEATH_RATE = float("{:.5f}".format(CAN_DEATH_RATE))

CAN_YTD_ACTUAL = int(yesterday['CANADA ACTUAL'])
CAN_DAY_DELTA = int(ACTUAL_CANADA)-int(CAN_YTD_ACTUAL)
CAN_DELTA_YTDY=int(yesterday['CAN DAY DELTA'])
CAN_INF_PT = int(CAN_DAY_DELTA)/int(CAN_DELTA_YTDY)

CAN_NAT_LOG = np.log(int(ACTUAL_CANADA)/int(CAN_YTD_ACTUAL))
CAN_NAT_LOG = float("{:.6f}".format(CAN_NAT_LOG))


ACTUAL_ONT = input("Ontario ")
ONT_YTD_ACTUAL = int(yesterday['ONTARIO'])
ONT_DAY_DELTA = int(ACTUAL_ONT)-int(ONT_YTD_ACTUAL)
ONT_DELTA_YTDY=int(yesterday['ONT DAY DELTA'])
ONT_INF_PT = int(ONT_DAY_DELTA)/int(ONT_DELTA_YTDY)
ONT_NAT_LOG = np.log(int(ACTUAL_ONT)/int(ONT_YTD_ACTUAL))
ONT_NAT_LOG = float("{:.4f}".format(ONT_NAT_LOG))

ACTUAL_NS = input ("NS ")

d = {
     'USA ACTUAL': pd.Series([ACTUAL_USA], index=[DATE]),
     'USA DEAD': pd.Series([ACTUAL_USA_DEAD], index=[DATE]),
     'USA DEATH RATE': pd.Series((USA_DEATH_RATE), index=[DATE]),
     'USA NAT LOG': pd.Series((USA_NAT_LOG), index=[DATE]),
     'CANADA ACTUAL': pd.Series([ACTUAL_CANADA], index=[DATE]),
     'CANADA DEAD': pd.Series([ACTUAL_CAN_DEAD], index=[DATE]),
     'CANADA DEATH RATE': pd.Series((CAN_DEATH_RATE), index=[DATE]),
     'CANADA DAY DELTA': pd.Series((CAN_DAY_DELTA), index=[DATE]),
     'CANADA INF PT': pd.Series((CAN_INF_PT), index=[DATE]),
     'CANADA NAT LOG': pd.Series((CAN_NAT_LOG), index=[DATE]),
     'ONTARIO': pd.Series([ACTUAL_ONT], index=[DATE]),
     'ONTARIO DAY DELTA': pd.Series([ONT_DAY_DELTA], index=[DATE]),
     'ONTARIO INFL PNT': pd.Series([ONT_INF_PT], index=[DATE]),
     'ONT NAT LOG': pd.Series([ONT_NAT_LOG], index=[DATE]),
     'NS':pd.Series([ACTUAL_NS], index=[DATE])
}
                               
write_to_ontario_file = {
    'ONTARIO': pd.Series([ACTUAL_ONT], index=[DATE])
}   

write_to_usa_file = {
    'USA': pd.Series([ACTUAL_USA], index=[DATE])
}   

write_to_canada_file = {
    'CANADA': pd.Series([ACTUAL_CANADA], index=[DATE])
}   
                                  
                               

df =pd.DataFrame(d)
filename=DATE+'.csv'
df.to_csv(filename)                        
with open('Output.csv', 'a') as f:
    df.to_csv(f, header=False)

df_for_ontario_only = pd.DataFrame(write_to_ontario_file)                             
with open('python_csv/Ontario.csv', 'a') as g:
    df_for_ontario_only.to_csv(g, header=False)
    
df_for_usa_only = pd.DataFrame(write_to_usa_file)                             
with open('python_csv/USA.csv', 'a') as h:
    df_for_usa_only.to_csv(h, header=False)
    
df_for_canada_only = pd.DataFrame(write_to_canada_file)                             
with open('python_csv/Canada.csv', 'a') as i:
    df_for_canada_only.to_csv(i, header=False)
                               

# diff_cases_USA=Cases_Usa_P-Cases_Usa_A
# diff_deaths_USA=Deaths_Usa_P-Deaths_Usa_A
# diff_Canada_cases= Cases_Canada_P-Cases_Canada_A
# diff_Canada_deaths= Deaths_Canada_P - Deaths_Canada_A
# diff_Ontario_cases = Ontario_Cases_P-Ontario_Cases_A




print("===============================")
print("Usa Actual Yesterday:", yesterday['USA ACTUAL'])
print("Usa Dead Yesterday: ", yesterday['USA DEAD'])
print("Canada Actual Yesterday: ", yesterday['CANADA ACTUAL'])
print("Candad Dead Yesterday: ", yesterday['CANADA DEAD'])
                               
```
